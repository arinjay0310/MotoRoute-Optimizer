```python
class Town:
    """Represents a Node in our graph."""
    def __init__(self, name, has_fuel_station, elevation_meters=0):
        self.name = name
        self.has_fuel_station = has_fuel_station
        self.elevation = elevation_meters

    def __repr__(self):
        fuel_status = "⛽ Yes" if self.has_fuel_station else "🚫 No"
        return f"{self.name} (Fuel: {fuel_status}, Elev: {self.elevation}m)"


class RouteMap:
    """Represents the Graph using an adjacency list."""
    def __init__(self):
        # Dictionary to store Town objects: { "TownName": TownObject }
        self.towns = {}
        # Adjacency list: { "TownName": { "DestinationName": {"distance": X, "time": Y} } }
        self.graph = {}

    def add_town(self, name, has_fuel_station=False, elevation_meters=0):
        """Adds a new town (vertex) to the map."""
        if name not in self.towns:
            new_town = Town(name, has_fuel_station, elevation_meters)
            self.towns[name] = new_town
            self.graph[name] = {}
            print(f"Added town: {new_town}")

    def add_route(self, town_a, town_b, distance_km, time_hours, is_two_way=True):
        """Adds a road (edge) between two towns."""
        if town_a not in self.towns or town_b not in self.towns:
            raise ValueError("Both towns must be added to the map before creating a route.")

        # Add edge from A to B
        self.graph[town_a][town_b] = {
            "distance_km": distance_km,
            "time_hours": time_hours
        }

        # If it's a two-way mountain road, add the reverse edge
        if is_two_way:
            self.graph[town_b][town_a] = {
                "distance_km": distance_km,
                "time_hours": time_hours
            }
        print(f"Added route: {town_a} <--> {town_b} ({distance_km}km, {time_hours}hrs)")

    def display_map(self):
        """Prints the current state of the graph."""
        print("\n--- Current Route Map ---")
        for town, connections in self.graph.items():
            print(f"\n📍 {town}:")
            if not connections:
                print("  (No outgoing routes)")
            for dest, data in connections.items():
                print(f"  -> to {dest}: {data['distance_km']} km, {data['time_hours']} hrs")


# ==========================================
# Test out the data structure
# ==========================================
if __name__ == "__main__":
    # Initialize the graph
    himalayan_circuit = RouteMap()

    print("Initializing Data...\n")

    # 1. Add Nodes (Towns)
    himalayan_circuit.add_town("Shimla", has_fuel_station=True, elevation_meters=2276)
    himalayan_circuit.add_town("Narkanda", has_fuel_station=True, elevation_meters=2708)
    himalayan_circuit.add_town("Sangla", has_fuel_station=True, elevation_meters=2621)
    himalayan_circuit.add_town("Chitkul", has_fuel_station=False, elevation_meters=3450)

    print("\nMapping Routes...\n")

    # 2. Add Edges (Roads/Routes)
    himalayan_circuit.add_route("Shimla", "Narkanda", distance_km=60, time_hours=2.5)
    himalayan_circuit.add_route("Narkanda", "Sangla", distance_km=160, time_hours=5.5)
    himalayan_circuit.add_route("Sangla", "Chitkul", distance_km=28, time_hours=1.5)

    # 3. View the structure
    himalayan_circuit.display_map()
```
